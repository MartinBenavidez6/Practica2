```python
names = """ Agustin, Yanina, Andrés, Ariadna, Bautista, CAROLINA,
CESAR, David, Diego, Dolores, DYLAN, ELIANA, Emanuel, Fabián, Noelia,
Francsica', FEDERICO, Fernanda, GONZALO, Nancy """
goals = [0, 10, 4, 0, 5, 14, 0, 0, 7, 2, 1, 1, 1, 5, 6, 1, 1, 2, 0,
11]
goals_avoided = [0, 2, 0, 0, 5, 2, 0, 0, 1, 2, 0, 5, 5, 0, 1, 0, 2,
3, 0, 0]
assists = [0, 5, 1, 0, 5, 2, 0, 0, 1, 2, 1, 5, 5, 0, 1, 0, 2, 3, 1,
0]

names = names.lower()

def generar_estadisticas(names, goals, goals_avoided, assists):
    estadisticas = []
    for i in range(len(names)):
        jugador = {
            "Nombre": names[i].strip(),
            "Goles a favor": goals[i],
            "Goles evitados": goals_avoided[i],
            "Asistencias": assists[i]
        }
        estadisticas.append(jugador)
    return estadisticas

def obtener_goleador(estadisticas):
    max_goles = max(estadisticas, key=lambda x: x["Goles a favor"])
    return max_goles["Nombre"], max_goles["Goles a favor"]

def jugador_mas_influyente(estadisticas):
    max_influencia = max(estadisticas, key=lambda x: x["Goles a favor"] * 1.5 + x["Goles evitados"] * 1.25 + x["Asistencias"])
    return max_influencia["Nombre"]

def promedio_goles_equipo(goals):
    total_goles = sum(goals)
    partidos_jugados = 25
    promedio = total_goles / partidos_jugados
    return promedio

def promedio_goles_goleador(goles_goleador):
    partidos_jugados = 25
    promedio = goles_goleador / partidos_jugados
    return promedio

estadisticas_jugadores = generar_estadisticas(names.split(','), goals, goals_avoided, assists)

goleador_nombre, goleador_goles = obtener_goleador(estadisticas_jugadores)

jugador_influyente = jugador_mas_influyente(estadisticas_jugadores)

promedio_equipo = promedio_goles_equipo(goals)

promedio_goleador = promedio_goles_goleador(goleador_goles)

print(" 1. Estructura de estadísticas de cada jugador:")
for jugador in estadisticas_jugadores:
    print(jugador)

print("\n 2. Goleador de la temporada:")
print("Nombre:", goleador_nombre)
print("Goles:", goleador_goles)

print("\n 3. Jugador más influyente de la temporada:")
print("Nombre:", jugador_influyente)

print("\n 4. Promedio de goles por partido del equipo:")
print("Promedio:", promedio_equipo)

print("\n 5. Promedio de goles por partido del goleador:")
print("Promedio:", promedio_goleador)
```
